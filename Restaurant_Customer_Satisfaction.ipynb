# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.drop('HighSatisfaction',axis=1),data[['HighSatisfaction']],test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y_train
```

# Column Transformation

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('tnf1', StandardScaler(), ['Age', 'Income', 'AverageSpend', 'WaitTime', 'ServiceRating', 'FoodRating', 'AmbianceRating']),
        ('tnf2', OrdinalEncoder(categories=[['Female', 'Male'], ['Rarely', 'Daily', 'Weekly', 'Monthly'], ['American', 'Chinese', 'Indian', 'Italian', 'Mexican'], ['Breakfast', 'Lunch', 'Dinner'], ['Casual', 'Business', 'Celebration'], ['Dine-in', 'Takeaway']]), ['Gender', 'VisitFrequency', 'PreferredCuisine', 'TimeOfVisit', 'DiningOccasion', 'MealType'])
    ],
    remainder='passthrough' 
)
```

```python
X_train_transformed=transformer.fit_transform(X_train)
```

```python
X_train_transformed_df=pd.DataFrame(X_train_transformed)
```

```python
X_test_transformed=transformer.transform(X_test)
```

```python
X_test_transformed_df=pd.DataFrame(X_test_transformed)
```

# Model Selection 

```python
#  Logistic Regression, KNN, Decision Tree Classifier, XG Boost Classifier, Random Forest Classifier
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
knn=KNeighborsClassifier()
dtc=DecisionTreeClassifier()
```

```python
lr.fit(X_train_transformed_df,y_train)
```

```python
y_pred_lr=lr.predict(X_test_transformed_df)
```

```python
accuracy_lr=accuracy_score(y_test,y_pred_lr)
print('Accuracy Score of Logistic Regression: ',accuracy_lr)
```

```python
knn.fit(X_train_transformed_df,y_train)
```

```python
y_pred_knn=knn.predict(X_test_transformed_df)
```

```python
accuracy_knn=accuracy_score(y_test,y_pred_knn)
print('Accuracy Score of Logistic Regression: ',accuracy_knn)
```

```python
dtc.fit(X_train_transformed_df,y_train)
```

```python
y_pred_dtc=dtc.predict(X_test_transformed_df)
```

```python
accuracy_dtc=accuracy_score(y_test,y_pred_dtc)
print('Accuracy Score of Logistic Regression: ',accuracy_dtc)
```

```python
#Probability function for Logistic Regrssion: For Train Data
lr_proba_train = lr.predict_proba(X_train_transformed_df)
lr_proba_train
```

```python
#For Test Data:
lr_proba_test = lr.predict_proba(X_test_transformed_df)
lr_proba_test
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
lr_fpr,lr_tpr,threshold=roc_curve(y_train,lr_proba_train[:, 1])
auc_lr=auc(lr_fpr,lr_tpr)
plt.plot(lr_fpr, lr_tpr, linestyle='-',label= 'LogisticRegression (auc=%0.3f)' %auc_lr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
l_fpr,l_tpr,threshold=roc_curve(y_test,lr_proba_test[:, 1])
auc_l=auc(l_fpr,l_tpr)
plt.plot(l_fpr, l_tpr, linestyle='-',label= 'LogisticRegression (auc=%0.3f)' %auc_l)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print('Confusion Matrix For Logistic Regression:')
pd.DataFrame(confusion_matrix(y_test, y_pred_lr),columns=list(range(0,2)))
```

```python
from sklearn.metrics import classification_report
```

```python
print('Classification Report for Logistic Regression:')
print(classification_report(y_test,y_pred_lr))
```
